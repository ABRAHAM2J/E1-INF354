**9. Con PYTHON sin librerías construya los índices de al menos dos ciclos para un Split de Train 80 y test de 20, apliquelo en el dataset iris.**

```python
import random
import pandas as pd
from sklearn.datasets import load_iris

# Cargar el conjunto de datos Iris
iris = pd.read_csv("iris.csv")

# Extraer los datos y etiquetas
datos = iris.iloc[:, :-1].values
etiquetas = iris.iloc[:, -1].values

# Definir el tamaño de entrenamiento y prueba
tamano_entrenamiento = 0.8
tamano_prueba = 0.2

# Calcular el número total de muestras
total_muestras = len(datos)

for ciclo in range(2):
    indices = list(range(total_muestras))

    # Establecer una semilla para la aleatoriedad (puedes cambiarla según tus necesidades)
    random.seed(54 + ciclo - 1)

    # Mezclar de los índices
    random.shuffle(indices)

    # Número de muestras en cada conjunto de entrenamiento y prueba
    muestras_entrenamiento = int(total_muestras * tamano_entrenamiento)
    muestras_prueba = total_muestras - muestras_entrenamiento

    # Dividir los índices en conjuntos de entrenamiento y prueba (80% - 20%)
    indices_entrenamiento = indices[:muestras_entrenamiento]
    indices_prueba = indices[muestras_entrenamiento:]

    # Crear conjuntos de datos de entrenamiento y prueba a partir de los índices
    datos_entrenamiento = [datos[i] for i in indices_entrenamiento]
    etiquetas_entrenamiento = [etiquetas[i] for i in indices_entrenamiento]

    datos_prueba = [datos[i] for i in indices_prueba]
    etiquetas_prueba = [etiquetas[i] for i in indices_prueba]

    # información sobre los conjuntos de entrenamiento y prueba por ciclo
    print(40*' '+ "Ciclo " + str(ciclo + 1))
    print(80*'-')
    print("Tamaño del conjunto de entrenamiento:", len(datos_entrenamiento))
    print("Tamaño del conjunto de prueba:", len(datos_prueba))
    
    # Datos de entrenamiento y etiquetas
    print("Conjunto de entrenamiento:")
    for i in range(len(datos_entrenamiento)):
        print("Dato:", datos_entrenamiento[i], "Etiqueta:", etiquetas_entrenamiento[i])

    # Imprimir datos de prueba y etiquetas
    print("Conjunto de prueba:")
    for i in range(len(datos_prueba)):
        print("Dato:", datos_prueba[i], "Etiqueta:", etiquetas_prueba[i])
    
    print(80*'-')
    print()
```
